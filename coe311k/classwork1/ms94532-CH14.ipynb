```python
# Step size
#h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
#df_dx = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
#df_dx_forward = (f(x + h) - f(x)) / h

# Backward Difference Approximation
#df_dx_backward = (f(x) - f(x - h)) / h

import matplotlib.pyplot as plt

import numpy as np
leftBound = 0 
rightBound = 2 * np.pi
xVals = np.linspace(leftBound, rightBound, 12)

h = (rightBound - leftBound) / (len(xVals) -1 )
#Central Difference:
df_dx = (np.cos(xVals + h) - np.cos(xVals - h)) / (2*h)
#Forward Difference 
df_dx_forward = (np.cos(xVals+h) - np.cos(xVals)) / h
#Backward Difference
df_dx_backward = (np.cos(xVals) - np.cos(xVals-h)) / h

import matplotlib.pyplot as plt
##Chatgpt to help with PLOT formatting 
plt.subplot(2, 2, 1)
plt.plot(xVals, np.cos(xVals))
plt.title('cos(x) - True Derivative')

plt.subplot(2, 2, 2)
plt.plot(xVals, df_dx_forward)
plt.title('Forward Difference')

plt.subplot(2, 2, 3)
plt.plot(xVals, df_dx_backward)
plt.title('Backward Difference')

plt.subplot(2, 2, 4)
plt.plot(xVals, df_dx)
plt.title('Central Difference')

plt.tight_layout()
plt.show()
```
