```python
def isDeterminant(m):
    place = 0
    count = 0
    for i in range (len(m)):
        for j in range (len(m[0])):
            if (j != place):
                count = count + abs ( m[i][j] ) 
        if (abs (m[i][place] ) < count):
            return False
        count = 0
        place = place + 1
    
    return True
        
        
            
    
```
