```python
import random 
import matplotlib.pyplot as plt

def approximatePi(numSamples): ## Followed guideline on slides
    counter = 0
    for i in range(numSamples):
        x = random.uniform(-1,1) 
        y = random.uniform(-1,1)
        
        if (x * x + y * y < 1):
            counter = counter + 1 
            
    return 4 * counter / numSamples 

import math ##this is to utilize math.pi variable in the original HW

absValList = [] ##
relValList = []
for x in range(0, 101, 1): ##utilized Chatgpt to figure out why I was getting a divided by 0 error, did not know until now that python re-instantiates a variablee, even if it is already defined, when utilized in a for loop without parameters.
    print("Absolute Error")
    value = approximatePi(1000)
    absVal = round (  abs(value - math.pi), 15 )
    relVal = round ( abs(value - math.pi), 15) / math.pi 
    absValList.append(absVal)
    relValList.append(relVal)
    print(absVal )  ##W3schools to learn round() function
    print("Relative Error")
    print(relVal)
    
plt.plot(absValList, color="blue") ##Slides from graphing 
plt.show()
plt.plot(relValList, color="red")
plt.show()
    

```
