```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
k = 10.0        # Constant in the equation (e.g., spring constant)
h = .01       # Step size
x_start = 0    # Start of the interval
x_end = 10     # End of the interval
y_0 = 1.0      # Initial condition for y
v_0 = 0.0      # Initial condition for dy/dx

hVal = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]

plt.figure(figsize=(10, 5))


for h in hVal:
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))


    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0

    # Euler's Method loop
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]

        # Update equations
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)

    # Plot the results
    
    plt.plot(x_values, y_values, label=f'h = {h}') ##AI - for better graph
    
plt.xlabel('x')
plt.ylabel('y and v')
plt.title("Solution of d²y/dx² = -ky using Euler's Method")
plt.legend()
plt.grid(True)
plt.ylim(-10, 10)
plt.show()
```

The higher the "k", then the faster oscillations take place. The lower the "k" value, the fewer number of oscillatinos take place from x:(0,10)

With a larger "h" value, Euler's Method may not be able to solve it becauese the step value would be far too great. By attempting. k = 10 and h = 10, this yielded a line that was obviously incorrrect. The larger the "k" value, the smaller the "h" values has to be. 

The graph becomes unstable when the "h" value exceeds the acceptable for the oscillation frequency. It seems that "k" and "h" have to be inverse in terms of how large their values are. The larger the "k" the smaller the "h" must be in order for the graph to remain acceptable.
