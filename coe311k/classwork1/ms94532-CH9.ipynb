```python
import numpy as np
import matplotlib

# Data points

X = np.array([1, 2, 3, 4, 5])

Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack the design matrix with a column of ones (for the intercept)

A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")

import matplotlib.pyplot as plt

plt.scatter(X, Y)

plt.plot(X, m*X + c) ##Chatgpt to line of best fit
plt.xlabel("X Axis")


plt.ylabel("Y. Axis")

plt.show()


X2 = np.random.rand(10)
Y2 = np.random.rand(10)
m2 = np.random.rand() ##thought wrong

c2 = np.random.rand()

A2 = np.vstack([X2, np.ones(len(X2))]).T
m2, c2 = np.linalg.lstsq(A2, Y2, rcond=None)[0]

plt.scatter(X2, Y2)


plt.plot(X2, m2*X2 + c2)
plt.show()



##Answers do not fit at all now as the Line of best fit is not calibrated to the randomized points


#Part 3
part3Slope = 4
part3Intercept = 2

X = np.arange(0, 1, 0.01)

Y = part3Slope*X + part3Intercept + (np.random.rand(len(X)) - 0.5)*0.2

A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
plt.scatter(X, Y)

plt.plot(X, part3Slope*X + part3Intercept)
plt.plot(X, m*X + c)


plt.show()
```
