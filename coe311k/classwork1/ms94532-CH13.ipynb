```python
import numpy as np ##From lectures 1:1 
import matplotlib.pyplot as plt
import time


def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    

    return C






    


```

```python
i = 3
naive_times = []
numpy_times = []
sizes = []
while(i <= 384):
    
    print("i = " + str(i))
    a = np.random.rand(i,i)
    b = np.random.rand(i,i)
  
    start = time.time()
    naive_matrix_multiplication(a.tolist(), b.tolist())
    end = time.time()
    naive_times.append(end - start)

    
    print("naive time is: " + str(end-start))
    start = time.time()
    np.dot(a, b)
    end = time.time()
    numpy_times.append(end - start)
    sizes.append(i)
    print("numpy time is: " + str(end-start))
    i = i * 2
    
    
    
plt.plot(sizes, naive_times, label='Naive')
plt.plot(sizes, numpy_times, label='NumPy')

plt.xlabel('Matrix Size (n)')
plt.ylabel('Runtime (seconds)')
plt.title('Matrix Multiplication Runtime')

plt.legend()
plt.show()
   

    



    
    
   
```
