```python
import math
import matplotlib.pyplot as plt
##Taylor series 


def taylorSin(x, terms, count):
    
    if ( count is None):
        count = 0
    if ( 2*terms -1 - count * 2 <= 0):
        return 0
    if ((terms - count )% 2 == 0):
        return ( (-1)*( x ** ( 2 * terms - 1 - count * 2) ) / (math.factorial (terms * 2 -1 - count * 2)) ) + taylorSin(x, terms, count+1)
 
    elif ( ( terms - count ) % 2 != 0):
        return  ( ( x ** ( 2 * terms - 1 - count * 2) ) / (math.factorial (terms * 2 -1 - count * 2)) )  + taylorSin(x, terms, count+1)
        return  ( ( x ** ( 2* terms - 1 - count*2  ) ) / ( math.factorial(terms * 2 - 1 - count*2) ) ) + taylorSin(x, terms, count+1) 
    
    
    
```

```python
i = 1
abs1 = []
abs2 = []
abs3 = []
abs4 = []

domain = []
while (i < 10.1):
    for k in range (1, 5, 1):
        print(str(i) + " : " + str(taylorSin(i, k, 0)) )
        print("Abs Error: " + str ( abs( taylorSin(i, k, 0) - math.sin(i)   ) ))
        if (k == 1):
            abs1.append( abs ( taylorSin(i, k, 0) - math.sin(i) ) ) 
        elif (k == 2):
            abs2.append( abs ( taylorSin(i, k, 0) - math.sin(i) ) ) 
        elif(k == 3):
            abs3.append( abs ( taylorSin(i, k, 0) - math.sin(i) ) ) 
        elif(k == 4):
            abs4.append( abs ( taylorSin(i, k, 0) - math.sin(i) ) ) 
    i = i + .1
    domain.append(i)
    
plt.plot(domain, abs1, color="maroon")

plt.plot(domain, abs2, color="blue")

plt.plot(domain, abs3, color="green")

plt.plot(domain, abs4, color="purple")
plt.show()
```
