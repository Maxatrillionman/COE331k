```python
import random 
def approximatePi(numSamples): ## Followed guideline on slides
    counter = 0
    for i in range(numSamples):
        x = random.uniform(-1,1) 
        y = random.uniform(-1,1)
        
        if (x * x + y * y < 1):
            counter = counter + 1 
            
    return 4 * counter / numSamples 
        
```
