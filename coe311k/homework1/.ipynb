```python
import numpy as np 

#Part 3 had AI help a lot, was confused but understand it now, issue was finding index of nearest value to x = 15
import matplotlib.pyplot as plt ##Missed this class day, AI helped me understand explanature. 
## From Notes 
def f(x, y):
    return -2 * y  * np.cos(x)
    ##Used debugging function, suh as floats not acceptable in for loop
# Define parameters

h_values = [0.01]
yTest = []
averageError = [] 
for i in range(5, 105, 5):
    h_values.append(i / 100)

plt.figure(figsize=(14, 6)) ##AI used to clean up graph
    
x_start, x_end = 0, 10 * np.pi

for h in h_values:
    num_steps = int((x_end - x_start) / h)

    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 1  # y(0) = 1

    # Function representing dy/dx = -2y * cos(x) , y(0) = 1 


    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
    y = np.exp(-2 * np.sin(x_values))
    averageError.append( np.mean( np.abs(y - y_values) )) ##AI used, debugging + np.mean was dividing instead of using the built in command which was yielding the wrong answer 
    # Plot the results

    yTrue = np.exp(-2*np.sin(x_values[np.argmin(np.abs( x_values - 15 )) ]))
    locError = np.abs(y_values[ np.argmin(np.abs(x_values - 15)) ] - yTrue )
    yTest.append(locError)
    plt.plot(x_values, y_values, label=f"h={h:.2f}") ##used to clean up grpah 
plt.xlabel('x')
plt.ylabel('y')
plt.title("Approximation of dy/dx = -2y using Euler's Method")
plt.legend()
plt.show()

plt.figure(figsize=(8, 5)) ##AI used for graph help 
plt.plot(h_values, averageError, marker='o')
plt.xlabel('h')
plt.ylabel('Average Error')
plt.title('Average Error against step size of h')
plt.show()

plt.figure(figsize=(8, 5)) ##AI used for graph help 
plt.plot(h_values, yTest, marker='o')
plt.xlabel('h')
plt.ylabel('Loc Error at x about 15')
plt.title('Lc Error at about x = 15 against step size of h')
plt.show()
```

What we are seeing here initially is the Euler's method approximation of the fucntion dy/dx = -2y*cos(x). This approximation is done seperately for each iteraiton of the step size. The subsequent graphs compare the each step sizes average error and loal error against step size of h. 

From the graphs it is clear that the smaller step sizes then the more accurate the approximation is. This is because more accuracy is cut-off with the larger step values (more straight lines cutting off accuracy)

The rigid shape of the third graph is because some functinons are closer to x = 15 than others, and this accounts for the seemingly random difference in local error despite the change in h between x = .2 and .8.
