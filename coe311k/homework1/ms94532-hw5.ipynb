```python
def leastSquare (vector1, vector2): ##My equation; used Chatgpt to find notation bug (.length error)
    
    xtotal = 0 
    ytotal = 0
    xy = 0 
    xSquared = 0

    
    for i in range (len(vector1)):
        xtotal = xtotal + vector1[i]
        ytotal = ytotal + vector2[i]
        
        xy = xy + vector1[i] * vector2[i]
        xSquared = xSquared + (vector1[i] * vector1[i])
    
    return ((len(vector1) * (xy) - (xtotal)*(ytotal)) / (len(vector1) * xSquared - (xtotal * xtotal)))
```

```python
Slope = 8
Intercept = 4

X = np.arange(0, 1.01, 0.01)

X = np.sort(X)

Y = Slope*X + Intercept + np.random.rand(len(X))

mySlope = leastSquare(X, Y)

myIntercept = np.mean(Y) - mySlope*np.mean(X)

print("My function ")
print("Slope: ")
print(mySlope)
print("Intercept:")
print(myIntercept)

A = np.vstack([X, np.ones(len(X))]).T ##Chatgpt utilized layout and organization 
m, c = np.linalg.lstsq(A, Y, rcond=None)[0] ##

print("Numpy Least Squares")
print("Slope:")
print(m)
print("Intercept:")
print(c)

predictY = mySlope*X + myIntercept

resid = Y - predictY

print("Residuals:")
print(resid)

largest = np.argmax(np.abs(resid))
smallest = np.argmin(np.abs(resid))

print("Largest residual point:")
print(X[largest], Y[largest])

print("Smallest residual point:")
print(X[smallest], Y[smallest])

plt.scatter(X, Y)

plt.plot(X, predictY) ##Chatgpt to aide plot 

plt.plot(X, m*X + c)

plt.show()
```
