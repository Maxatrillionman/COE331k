```python
import numpy as np

def jacobi(A, b, x0, tol, max_iterations):
    absError = []
    n = len(b)
    x = x0.copy()

    for k in range(max_iterations):
        x_new = np.zeros_like(x)

        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
            
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k

        x = x_new

    return x, max_iterations
```

```python
import numpy as np

def gauss_seidel(A, b, x0, tol, max_iterations):
    absError = []
    
    n = len(b)
    x = x0.copy()

    for k in range(max_iterations):
        x_new = x.copy()

        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]

        

        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k

        
        x = x_new
        
        

    return x, max_iterations
    
```

```python
A = np.array([
    [10, -1,  2,  0,  0],
    [-1, 11, -1,  3,  0],
    [ 2, -1, 10, -1,  0],
    [ 0,  3, -1,  8, -2],
    [ 0,  0,  0, -2,  9]
], dtype=float)


b = np.array([
    14,
    30,
    26,
    25,
    37
], dtype=float)

x0 = np.zeros_like(b)
x, max_iterations = jacobi(A, b, x0, 1e-6, 102)

print("Jacobi")
print(x)

print("Number of iterations: ")
print(max_iterations)

print(" ")
print("Gauss-seidel")
x, max_iterations = gauss_seidel(A, b, x0, 1e-6, 102)
print(x)

print("Number of iterations: ")
print(max_iterations)
```
